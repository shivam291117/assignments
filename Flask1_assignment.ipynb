**Q1. What is Flask Framework? What are the advantages of Flask Framework?**

Ans. Flask is a lightweight and micro web framework for Python that is designed to make it easy to start with web development quickly and with minimal overhead. Flask is classified as a microframework because it does not include many built-in tools or libraries that full-stack frameworks like Django might provide. Instead, it offers the essentials for building web applications and allows developers to extend its functionality with additional modules or plugins as needed.

**Advantages of Flask Framework :**
1. Simplicity and Ease of Use
2. Flexibility and Freedom
3. Lightweight
4. Modular and Extensible

**Q3. What is App routing in Flask? Why do we use app routes?**

Ans . App routing in Flask is the mechanism used to map URLs to specific functions within your web application. This is achieved using the @app.route decorator, which associates a particular URL path with a view function that handles requests to that path. The view function processes the request and returns a response, such as an HTML page or JSON data.

App routes are essential for defining how users interact with the web application. Here are some reasons why app routes are used in Flask:
1. URL Mapping
2. Handling Different HTTP Methods
3. Dynamic URL Building
4. Organizing Views

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.**

Ans. In Flask, the url_for() function is used to build a URL to a specific function. This is particularly useful for generating URLs dynamically and ensuring that they are correctly constructed based on the application's routing configuration. It helps to avoid hardcoding URLs in the templates and views, making the application more maintainable and flexible.


```python
from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')
def index():
    welcome_url = url_for('welcome')
    return f'This is the home page. Go to the welcome page: <a href="{welcome_url}">{welcome_url}</a>'

@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

@app.route('/details')
def company_details():
    details = """
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999
    """
    return details

@app.route('/redirect_to_details')
def redirect_to_details():
    details_url = url_for('company_details')
    return redirect(details_url)

if __name__ == '__main__':
    app.run(debug=True)
```
