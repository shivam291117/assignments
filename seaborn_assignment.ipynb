Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans.

Seaborn is a powerful visualization library built on top of Matplotlib, providing a high-level interface for drawing attractive and informative statistical graphics. Here are five types of plots you can create using Seaborn and their uses:

1. Scatter Plot

Function: sns.scatterplot()
Use: Scatter plots are used to visualize the relationship between two continuous variables. They can highlight correlations, clusters, and outliers.

2. Bar Plot

Function: sns.barplot()
Use: Bar plots are used to display the distribution of categorical data. They show the central tendency of the data along with confidence intervals.

3. Box Plot

Function: sns.boxplot()
Use: Box plots (or box-and-whisker plots) are used to show the distribution of data and identify outliers. They provide a summary of the data's minimum, first quartile, median, third quartile, and maximum.

4. Heatmap

Function: sns.heatmap()
Use: Heatmaps are used to display the magnitude of values in a matrix as colors. They are useful for visualizing correlation matrices, missing data, or any other matrix data.

5. Violin Plot

Function: sns.violinplot()
Use: Violin plots are used to visualize the distribution of the data across different categories. They combine aspects of box plots and kernel density plots to show the density of the data at different values.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using

x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset('fmri')

sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')

plt.title("FMRI: Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using

x ='pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()

plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column.

Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', palette='Set2')

plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species', palette='Set2')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')

flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')

plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='YlGnBu')

plt.title('Heatmap of Monthly Passengers Over Years')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
