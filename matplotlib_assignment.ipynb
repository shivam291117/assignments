Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Ans.

Scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are coded, one additional variable can be displayed.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x, y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans.

The subplot() function in Matplotlib is used to create multiple plots within a single figure. This is useful for comparing different datasets side by side, or to display multiple visualizations in a structured and organized manner. By using subplot(), you can specify the layout of the plots (rows and columns) and control the position of each plot within the figure.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o')
plt.title("Line Plot 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='o')
plt.title("Line Plot 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='o')
plt.title("Line Plot 3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='o')
plt.title("Line Plot 4")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

**Ans.**

A bar plot (or bar chart) is a graphical representation of data where individual bars represent the quantities of different categories. The length of each bar corresponds to the value it represents. Bar plots can be plotted vertically or horizontally

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(company, profit, color='blue')
plt.title("Company Profit - Vertical Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")

plt.subplot(1, 2, 2)
plt.barh(company, profit, color='green')
plt.title("Company Profit - Horizontal Bar Plot")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans.

A box plot, also known as a whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It can also highlight outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(10, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title("Box Plot of Two Datasets")
plt.xlabel("Dataset")
plt.ylabel("Values")
plt.grid(True)

plt.show()
```
