**Q1. Explain GET and POST methods.**

Ans. The GET method is used to request data from a specified resource.

The POST method is used to send data to a server to create/update a resource.

**Q2. Why is request used in Flask?**

Ans. The request object in Flask is essential for handling incoming request data. It allows developers to access form data, query parameters, JSON data, headers, cookies, request methods, and uploaded files. This enables the creation of dynamic and interactive web applications that can respond to various types of user input and other request data.

**Q4. What are templates in Flask? Why is the render_template() function used?**

Ans. Templates in Flask are used to separate the logic of your application from the presentation layer. This means that you can create HTML files with placeholders for dynamic content, and Flask will replace those placeholders with actual data before rendering the HTML to the user.
